```python
companies = ['A','B', 'C','D' , 'E','F']
dates = ['Enero', 'Febrero', 'Marzo']
```

```python
n_companies = len(companies)
n_dates = len(dates)
n_char = 4
```

```python
chars = ['x_' + str(i) for i in range(n_char)]
```

```python
np.repeat(dates, n_char)
```

```python
dates_indices = np.tile(np.repeat(dates, n_char), n_companies)

print(len(dates_indices))
dates_indices
```

```python
char_indices = np.tile(chars, n_dates * n_companies)
print(len(char_indices))
char_indices
```

```python
company_indices = np.repeat(companies, n_dates * n_char)
print(len(company_indices))
company_indices
```

```python
len(data)
```

```python
df = pd.DataFrame({'Date': dates_indices,'Characteristic': char_indices,'Value': data, 'Stock': company_indices,})

df
```

```python
# Pivot DataFrame
df_pivot = df.pivot_table(index=['Stock', 'Date'],
                          columns='Characteristic',
                          values='Value')

# Reset index for a flat DataFrame
df_pivot.reset_index(inplace=True)

df_pivot
```
